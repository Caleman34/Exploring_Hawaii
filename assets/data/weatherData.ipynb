```python
import pandas as pd
import numpy as np

path = 'Hawaii_weather_data.csv'

df = pd.read_csv(path)
df.head(25)
```

```python
# get list of columns
df.columns
```

```python
# check for null values
df.isnull().sum()
```

```python
# some months for 2021 do not have data yet, so the zeros need to be dropped so the averages are not skewed when making calculations
df= df[df['Annual'] != 0]
df.head(25)
```

```python
# Export file as a CSV, Without the Pandas index and with the the header
df.to_csv("clean_weather_data.csv", index=False, header=True)
```

```python
# use groupby to group data by location
grouped_location_df = df.groupby(['Weather Type'])
grouped_location_df.head()
```
